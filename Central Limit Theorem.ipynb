```python
# Generate 2000 random numbers from a normal distribution with a mean of 0 and standard deviation of 1.
import numpy as np
import matplotlib.pyplot as plt

mean = 0
std_dev = 1

random_numbers = np.random.normal(mean,std_dev,2000)

# Plot a histogram of the random numbers with 200 bins to visualize the distribution.
plt.hist(random_numbers, bins=200)
plt.title("Distribution of random numbers")
plt.xlabel("Numbers")
plt.ylabel("Count")
plt.show()
```

```python
# For each sample, computing the mean and storing it in the list `means`.
means = []
for i in range(200):
 r = np.random.choice(random_numbers,10)
 mean1 = np.mean(r)
 means.append(mean1)
```

```python
plt.hist(means , bins= 'auto')
plt.title("Central Limit Theorem Visualization")
plt.xlabel("Sample Means")
plt.ylabel("Frequency ")
```

```python
#uniform distribution

uni = np.random.uniform(0, 1, size=2000)
plt.hist(uni , bins=20)
plt.title("Uniform Distribution")
plt.xlabel("Numbers with uniform distribution")
plt.ylabel("Frequency")
```

```python
means2 = []
for i in range(300):
 r2 = np.random.choice(uni,10)
 mean2 = np.mean(r2)
 means2.append(mean2)
```

```python
plt.hist(means2 , bins= 30)
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.title("Visualising the Central Limit Theorem")
```

```python
#exponential

exp = np.random.exponential(scale=1, size=200)
plt.hist(exp , 200)
plt.xlabel("Numbers with exponential distribution")
plt.ylabel("Count")
plt.title("Random numbers with exponential distribution")
```

```python
means3 = []
for i in range(300):
 r3 = np.random.choice(exp,10)
 mean3 = np.mean(r3)
 means3.append(mean3)
```

```python
plt.hist(means3 , bins= 30)
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.title("Visualising the Central Limit Theorem")
```
