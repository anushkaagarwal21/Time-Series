```python
t1 = data1[:,0]
f1 = data1[:,1]
err1 = np.sqrt(data1[:,1])
```

```python
plt.plot(t1, f1, "o")
plt.xlabel("Time")
plt.ylabel("Flux")
plt.grid(True)
```

```python
def straightline_model (x, m ,c):
  return m*x+c
```

```python
params , cov = curve_fit(straightline_model , t1 , f1)
```

```python
params
```

```python
plt.scatter(t1,f1, label='Data')
plt.plot(t1, straightline_model(t1, params[0], params[1]), color='red', label = 'Best fit')
plt.title("Time Series - Chi-Square Fitiing")
plt.xlabel("Time")
plt.ylabel("Flux")
plt.grid(True)

plt.show()

```

```python
plt.scatter(t1, f1, label = "Data")
plt.plot(t1,straightline_model(t1,params[0], params[1]), color='red', label = 'Best fit')
plt.errorbar(t1,f1,yerr= err1)
```

```python
residuals = f1 - straightline_model(t1, params[0], params[1])
```

```python
plt.hist(residuals)
plt.grid(True)
plt.show()
```
